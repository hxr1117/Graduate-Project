```python
"""
This function is to read log data from files and store them in ...
"""
def read_dataset(path, log_type):
    # get all file list in path
    for root, dirs, files in os.walk(path):
        all_file_list = files
        
    apache_file = []
    for file_name in all_file_list:
        # If it is "Error Log"
        if file_name.split(".")[0] == log_type:
            file = open(path+file_name, "r", encoding='iso-8859-1')
            apache_file.extend(file.readlines())
            
    return apache_file
```

```python
"""
Aim of this function is to pre-process data, like removing the time.
Only support Error Log temporarily
"""
def clean_data(data):
    # Delete consecutive duplicate logs
    # Cannot detect multiple events that repeat in sequence
    logs = []
    for i in range(len(data)):
        # Replace specific ip address with "IP" and same as the "path"
        tmp = re.sub(r"(\d{1,3}\.){3}\d{1,3}((:)\d{1,4})?$", "*", data[i])  # match IP address and port
        tmp = re.sub(r"^(.+)\/([^\/]+)$", "*", tmp)  # match file paths
        tmp = re.sub(r"(19|20)\d\d[- /.](0[1-9]|1[012])[- /.](0[1-9]|[12][0-9]|3[01])\S*", "*", tmp)  # match date
        if len(logs) != 0:
            # If the current log is the same as the previous one, it means there is a duplicate
            if logs[-1] != tmp:
                logs.append(tmp)
        else:
            logs.append(tmp)
    return logs
```

```python
"""
This function is aim to generate a n-gram array for each sentense.
"""
def generate_n_gram(data, ngram=1):
    terms = []
    for idx in range(len(data)):
        words = data[idx].split()
        temp = zip(*[words[i:] for i in range(0, ngram)])
        temp = [' '.join(ngram) for ngram in temp]
        data[idx] = temp
        terms.(temp)
    terms_list = set(terms)
    return data, terms_list
```

```python
"""
This function is aim to generate n-gram/words frequency matrix.
"""
def generate_frequency_matrix(data, terms):
    matrix = pd.DataFrame(columns=terms)
    for idx in range(len(data)):
        freq = dict()
        for w in data[idx]:
            if w not in freq:
                freq[w] = 1
            else:
                freq[w] += 1
        matrix = matrix.append(freq, ignore_index=True)
    matrix = matrix.fillna(0)
    return matrix
```

```python
def generate_tokens_position_matrix(data):
    max_len = max([len(i) for i in data])
    matrix = dict()
    for idx in range(len(data)):  # idx for the idx-th log sentence
        for i in range(len(data[idx])):  # i for the position
            token = data[idx][i]
            # "token" = [0, 0, 1, 5]
            if token in matrix:
                matrix[token][i] += 1
            else:
                matrix[token] = [0] * max_len
                matrix[token][i] += 1
    return matrix
```
